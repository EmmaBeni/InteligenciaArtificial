```python
# Definimos las probabilidades (Tablas de Probabilidad Condicional)
# P(R)
P_R = {True: 0.05, False: 0.95}

# P(C | R) -> Formato: P_C_dado_R[estado_router][estado_celular]
P_C_dado_R = {
    True:  {True: 0.95, False: 0.05}, # Si el router FALLA
    False: {True: 0.10, False: 0.90}  # Si el router está OK
}

# P(T | R) -> Formato: P_T_dado_R[estado_router][estado_tv]
P_T_dado_R = {
    True:  {True: 0.99, False: 0.01}, # Si el router FALLA
    False: {True: 0.05, False: 0.95}  # Si el router está OK
}
```

```python
# 1. Regla de la Cadena: P(R, C, T) = P(R) * P(C|R) * P(T|R)
def probabilidad_conjunta(r, c, t):
    """Calcula la probabilidad de una combinación específica de eventos."""
    return P_R[r] * P_C_dado_R[r][c] * P_T_dado_R[r][t]


# 2. Motor de Inferencia: Teorema de Bayes y Marginalización
def inferencia_router_fallando(celular_falla=None, tv_falla=None):
    """Calcula P(Router=True | Evidencia). Si no pasas evidencia, ignora la variable."""
    
    # Valores posibles para las variables booleanas
    estados = [True, False]
    
    numerador = 0.0    # Probabilidad conjunta P(Router=True, Evidencia)
    denominador = 0.0  # Probabilidad marginal P(Evidencia)
    
    # Exploramos todas las combinaciones posibles (Marginalización)
    for c in estados:
        for t in estados:
            # Si pasamos evidencia y no coincide con el bucle actual, lo saltamos
            if celular_falla is not None and c != celular_falla: continue
            if tv_falla is not None and t != tv_falla: continue
            
            # Sumamos al denominador (todas las realidades donde ocurre la evidencia)
            prob_actual_r_true = probabilidad_conjunta(True, c, t)
            prob_actual_r_false = probabilidad_conjunta(False, c, t)
            
            numerador += prob_actual_r_true
            denominador += (prob_actual_r_true + prob_actual_r_false)
            
    return numerador / denominador
```

```python
print("--- DIAGNÓSTICO DE WIFI ---")

# Escenario 0: Sin saber nada (Probabilidad a priori)
prob_base = P_R[True]
print(f"0. A priori, la probabilidad de que el router falle es: {prob_base*100:.1f}%")

# Escenario 1: Solo el celular falla
prob_1 = inferencia_router_fallando(celular_falla=True)
print(f"1. Si el CELULAR falla, la prob. de que el router esté roto sube a: {prob_1*100:.1f}%")

# Escenario 2: El celular y la TV fallan
prob_2 = inferencia_router_fallando(celular_falla=True, tv_falla=True)
print(f"2. Si CELULAR Y TV fallan, la prob. de que el router esté roto salta a: {prob_2*100:.1f}%")
```
