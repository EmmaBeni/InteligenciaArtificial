```python
#1 Normalizar texto. Pasar a minúsuclas y eliminar puntuación

import re



def my_normalize(text):
    return ( re.sub(r"[^\w\sáéíóúüñÁÉÍÓÚÜÑ]", "", text)).lower()

# cualquier cosa de text que no concuerde con lo que hay en [], reemplazalo por ""
# r es para no tener que susar \\ y poder tenerla simple \

print(my_normalize("¡Buenos días, Emma! Espero estés bien."))
```

```python
#2 Tokenizar. Dividir el resultado del anterior

def tokenize(text):
    words = my_normalize(text)
    return words.split(" ")

```

```python
from nltk.corpus import stopwords


#3 Filtrar stopwords.

def my_filter(text):
    stop_words = stopwords.words('spanish')
    words = tokenize(text)
    return [word for word in words if not word in stop_words]

print(my_filter("¡Buenos días, Emma! Espero estés bien."))
```

```python
#4 Contabilizar cada palabra aparecida en todos los textos

def build_vocab(texts):
    vocabulary = {}
    index = 0

    for text in texts:
        tokens = my_filter(text)   # ya tokenizado y sin stopwords
        for token in tokens:
            if token not in vocabulary:
                vocabulary[token] = index
                index += 1
    return vocabulary
```

```python
#5 COnstruir vector

def vectorize(text, vocabulary):
    # 1. Obtener los tokens limpios del texto
    tokens = my_filter(text)

    # 2. Crear un vector de ceros del tamaño del vocabulario
    vector = [0] * len(vocabulary)

    # 3. Recorrer cada token del documento
    for token in tokens:
        if token in vocabulary:
            index = vocabulary[token]   # índice asignado en el paso 4
            vector[index] += 1          # incrementar la cuenta

    return vector
```

```python
#6 Matriz, vector de cada texto

def vectorize_all(texts, vocabulary):
    return [vectorize(text, vocabulary) for text in texts]
```

```python
def vectorize_sparse(texts, vocabulary):
    rows = []
    cols = []
    data = []

    for row_index, text in enumerate(texts):
        tokens = my_filter(text)

        # Contar ocurrencias en este documento
        counts = {}
        for token in tokens:
            if token in vocabulary:
                idx = vocabulary[token]
                counts[idx] = counts.get(idx, 0) + 1

        # Guardar en formato disperso
        for col_index, count in counts.items():
            rows.append(row_index)
            cols.append(col_index)
            data.append(count)

    return rows, cols, data
```
